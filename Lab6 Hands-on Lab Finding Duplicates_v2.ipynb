### Task 2: Analyze Characteristics of Duplicates
1. Identify duplicate rows based on selected columns such as MainBranch, Employment, and RemoteWork. Analyse which columns frequently contain identical values within these duplicate rows.
2. Analyse the characteristics of rows that are duplicates based on a subset of columns, such as MainBranch, Employment, and RemoteWork. Determine which columns frequently have identical values across these rows.
   


```python
## Write your code here

# Identify duplicate rows based on selected columns
subset_columns = ['MainBranch', 'Employment', 'RemoteWork']
subset_duplicates = df[df.duplicated(subset=subset_columns, keep=False)]

# Analyze frequently identical columns within duplicates
identical_columns = {}
if not subset_duplicates.empty:
    first_duplicate = subset_duplicates.iloc[0]
    for column in df.columns:
        if (subset_duplicates[column] == first_duplicate[column]).all():
            identical_columns[column] = True

print(f"Duplicate rows based on {subset_columns}:\n")
print(subset_duplicates.head().to_markdown(index=False, numalign="left", stralign="left"))

print("\nColumns with identical values in these duplicates:")
for col in identical_columns:
    print(f"- {col}")
```

### Task 3: Visualize Duplicates Distribution
1. Create visualizations to show the distribution of duplicates across different categories.
2. Use bar charts or pie charts to represent the distribution of duplicates by Country and Employment.


```python
!pip install seaborn
```

```python
## Write your code here

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset directly from the URL
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/VYPrOu0Vs3I0hKLLjiPGrA/survey-data-with-duplicate.csv"
df = pd.read_csv(file_path)

# Create a 'Duplicate' column to easily identify duplicates
df['Duplicate'] = df.duplicated(keep=False)

# Visualize duplicates by Country
plt.figure(figsize=(12, 6))
sns.countplot(y='Country', data=df[df['Duplicate']], order=df[df['Duplicate']]['Country'].value_counts().index)
plt.title('Distribution of Duplicates by Country')
plt.show()

# Visualize duplicates by Employment
plt.figure(figsize=(10, 5))
sns.countplot(y='Employment', data=df[df['Duplicate']], order=df[df['Duplicate']]['Employment'].value_counts().index)
plt.title('Distribution of Duplicates by Employment')
plt.show()

# Pie chart for Employment Duplicates (alternative visualization)
employment_duplicates = df[df['Duplicate']]['Employment'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(employment_duplicates, labels=employment_duplicates.index, autopct='%1.1f%%', startangle=140)
plt.title('Employment Duplicates Distribution (Pie Chart)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Pie chart for Country Duplicates (alternative visualization)
country_duplicates = df[df['Duplicate']]['Country'].value_counts()
if len(country_duplicates) <= 10: # avoid too many slices.
    plt.figure(figsize=(8, 8))
    plt.pie(country_duplicates, labels=country_duplicates.index, autopct='%1.1f%%', startangle=140)
    plt.title('Country Duplicates Distribution (Pie Chart)')
    plt.axis('equal')
    plt.show()
else:
    print("Too many countries for a pie chart, consider a bar chart.")
```

### Task 4: Strategic Removal of Duplicates
1. Decide which columns are critical for defining uniqueness in the dataset.
2. Remove duplicates based on a subset of columns if complete row duplication is not a good criterion.


```python
## Write your code here


# Assuming 'ResponseId' is a unique identifier (if it exists)
unique_columns = ['ResponseId']

# Check if 'ResponseId' exists, if not, use other relevant columns
if 'ResponseId' not in df.columns:
    unique_columns = ['MainBranch', 'Employment', 'Country', 'Age', 'EdLevel', 'RemoteWork', 'DevType'] #Add other relevant columns

# Remove duplicates based on the selected unique columns
df_unique = df.drop_duplicates(subset=unique_columns, keep='first')

# Verify the removal of duplicates
print(f"Original DataFrame shape: {df.shape}")
print(f"DataFrame after removing duplicates based on {unique_columns}: {df_unique.shape}")

# Verify that duplicates are gone.
print(f"Number of duplicate rows after removal: {df_unique.duplicated(subset=unique_columns).sum()}")
```
