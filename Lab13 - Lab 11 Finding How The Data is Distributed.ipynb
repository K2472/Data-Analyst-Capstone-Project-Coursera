### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
!pip install matplotlib-venn
```

```python
## Write your code here

# import pandas as pd
# import matplotlib.pyplot as plt
from matplotlib_venn import venn2  # Install: pip install matplotlib-venn

# 1. Load your data (replace 'your_dataset.csv' with your actual file)
# 2. Analyze Programming Languages

if 'LanguageHaveWorkedWith' in df.columns and 'LanguageWantToWorkWith' in df.columns:
    print("\n--- Analysis of Programming Languages ---")

    # A. Data Preparation (Important!)

    # The language columns likely contain multiple languages separated by semicolons.
    # We need to split these into lists of individual languages.

    def split_languages(language_string):
        if isinstance(language_string, str):  # Handle potential NaN values
          return [lang.strip() for lang in language_string.split(';')]
        return [] # Return empty list if language_string is not a string

    df['LanguageHaveWorkedWith'] = df['LanguageHaveWorkedWith'].apply(split_languages)
    df['LanguageWantToWorkWith'] = df['LanguageWantToWorkWith'].apply(split_languages)


    # Create sets of languages
    languages_worked_with = set()
    languages_want_to_work_with = set()

    for index in df.index:
        languages_worked_with.update(df['LanguageHaveWorkedWith'][index])
        languages_want_to_work_with.update(df['LanguageWantToWorkWith'][index])

    # Calculate frequencies
    worked_with_counts = {}
    want_to_work_with_counts = {}

    for lang in languages_worked_with:
        worked_with_counts[lang] = 0
    for lang in languages_want_to_work_with:
        want_to_work_with_counts[lang] = 0

    for index in df.index:
        for lang in df['LanguageHaveWorkedWith'][index]:
            if lang in worked_with_counts:
                worked_with_counts[lang] += 1
        for lang in df['LanguageWantToWorkWith'][index]:
            if lang in want_to_work_with_counts:
                want_to_work_with_counts[lang] += 1
    
    # Convert dictionaries to dataframes for easier handling
    worked_with_df = pd.DataFrame(list(worked_with_counts.items()), columns=['Language', 'WorkedWithCount'])
    want_to_work_with_df = pd.DataFrame(list(want_to_work_with_counts.items()), columns=['Language', 'WantToWorkWithCount'])

    # Merge the two dataframes on the 'Language' column using an 'outer' join
    languages_df = pd.merge(worked_with_df, want_to_work_with_df, on='Language', how='outer')

    # Fill NaN values with 0 (for languages that are only in one list)
    languages_df.fillna(0, inplace=True)

    # Convert counts to integers
    languages_df['WorkedWithCount'] = languages_df['WorkedWithCount'].astype(int)
    languages_df['WantToWorkWithCount'] = languages_df['WantToWorkWithCount'].astype(int)

    # B. Venn Diagram

    # Get the sets of languages
    worked_with = set(languages_df['Language'][languages_df['WorkedWithCount'] > 0])
    want_to_work_with = set(languages_df['Language'][languages_df['WantToWorkWithCount'] > 0])

    plt.figure(figsize=(8, 6))
    v = venn2([worked_with, want_to_work_with], ('Languages Worked With', 'Languages Wanted'))

    # Annotate the Venn diagram with counts (optional)
    v.get_label_by_id('100').set_text(len(worked_with - want_to_work_with))  # Only worked with
    v.get_label_by_id('010').set_text(len(want_to_work_with - worked_with))  # Only want to work with
    v.get_label_by_id('110').set_text(len(worked_with & want_to_work_with))  # Overlap

    plt.title("Overlap of Programming Languages")
    plt.show()


    # C. Grouped Bar Chart

    # Sort by total count for better visualization
    languages_df['Total'] = languages_df['WorkedWithCount'] + languages_df['WantToWorkWithCount']
    languages_df = languages_df.sort_values(by='Total', ascending=False)
    languages_df = languages_df.head(10) # Show top 10 languages


    plt.figure(figsize=(12, 6))  # Adjust figure size for more languages

    bar_width = 0.35
    index = range(len(languages_df))

    plt.bar(index, languages_df['WorkedWithCount'], bar_width, label='Worked With')
    plt.bar([i + bar_width for i in index], languages_df['WantToWorkWithCount'], bar_width, label='Want To Work With')

    plt.xlabel("Programming Languages")
    plt.ylabel("Frequency")
    plt.title("Frequency of Programming Languages (Grouped Bar Chart)")
    plt.xticks([i + bar_width/2 for i in index], languages_df['Language'], rotation=45, ha='right')  # Rotate x-axis labels
    plt.legend()
    plt.tight_layout() # Adjust layout to prevent labels from overlapping
    plt.show()

else:
    print("\nOne or both of the 'LanguageHaveWorkedWith' or 'LanguageWantToWorkWith' columns were not found.")
```

```python
#
#
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# 2. Analyze Remote Work Trends

if 'RemoteWork' in df.columns and 'Country' in df.columns:
    print("\n--- Analysis of Remote Work Trends ---")

    # Check for Missing Values (Important!)
    print(f"\nMissing Values in RemoteWork: {df['RemoteWork'].isnull().sum()}")
    print(f"Missing Values in Country: {df['Country'].isnull().sum()}")

    # Handle Missing Values (if needed - choose ONE method for each column)
    df['RemoteWork'] = df['RemoteWork'].fillna(df['RemoteWork'].mode()[0])  # Fill with mode.
    df['Country'] = df['Country'].fillna('Unknown')  # fill country with unknown

    # Re-check Missing Values after handling
    print(f"\nMissing Values in RemoteWork (After Handling): {df['RemoteWork'].isnull().sum()}")
    print(f"Missing Values in Country (After Handling): {df['Country'].isnull().sum()}")

    # A. Grouped Bar Chart
    top_n = 10  # Display top 10 countries
    top_countries = df['Country'].value_counts().nlargest(top_n).index
    remote_work_by_country = df[df['Country'].isin(top_countries)].groupby(['Country', 'RemoteWork']).size().unstack()

    plt.figure(figsize=(12, 6))
    remote_work_by_country.plot(kind='bar', figsize=(12, 6))
    plt.title(f"Remote Work Distribution by Top {top_n} Countries (Grouped Bar Chart)")
    plt.xlabel("Country")
    plt.ylabel("Frequency")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # B. Heatmap (Alternative Visualization)

    # Convert RemoteWork to numerical values for heatmap (if it's not already)
    if df['RemoteWork'].dtype == 'object':
        remote_work_mapping = {'Yes': 1, 'No': 0, 'Sometimes': 0.5}
        df['RemoteWork_Numerical'] = df['RemoteWork'].map(remote_work_mapping)

        remote_work_by_country_heatmap = df[df['Country'].isin(top_countries)].groupby('Country')['RemoteWork_Numerical'].mean()

        # Check for NaN values
        if remote_work_by_country_heatmap.isnull().all():
            print("Warning: All values in remote_work_by_country_heatmap are NaN. Skipping heatmap.")
        else:
            plt.figure(figsize=(10, 8))
            sns.heatmap(remote_work_by_country_heatmap.to_frame(), annot=True, cmap="YlGnBu", cbar_kws={'label': 'Average Remote Work'})
            plt.title(f"Remote Work Distribution by Top {top_n} Countries (Heatmap)")
            plt.xlabel("Remote Work (Average)")
            plt.ylabel("Country")
            plt.tight_layout()
            plt.show()

        del df['RemoteWork_Numerical']

    else:
        print("RemoteWork is already numerical. Skipping heatmap generation.")

else:
    print("\nOne or both of the 'RemoteWork' or 'Country' columns were not found.")
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)

```python
## Write your code here

import pandas as pd
import matplotlib.pyplot as plt

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Handle missing values (fill with mode)
if 'EdLevel' in df.columns:
    df['EdLevel'] = df['EdLevel'].fillna(df['EdLevel'].mode()[0])
else:
    print("EdLevel column not found.")

if 'Employment' in df.columns:
    df['Employment'] = df['Employment'].fillna(df['Employment'].mode()[0])
else:
    print("Employment column not found.")

# Create cross-tabulation
cross_tab = pd.crosstab(df['Employment'], df['EdLevel'])

# Print the cross-tabulation
print("Cross-tabulation of Employment vs. Education Level:\n", cross_tab)

# Visualize with a stacked bar plot
cross_tab.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Employment vs. Education Level')
plt.xlabel('Employment Status')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
